```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("sea_ice_extent_Dec_Northern_Hemisphere.csv")
df3 = pd.read_csv("cyclones_fig-3.csv")
df3.head()
df3.dropna()
```

```python
df.head()
```

```python
year = df["Date"]
extent = df["Extent (million km2)"]
anomaly = df["Anomaly (million km2)"]
year3 = df3["Year"]
sea_temp = df3["Smoothed sea surface temperature"]
```

```python
# cleaning up the graph
plt.clf()

# plotting the points; determining axes
plt.plot(year, anomaly, color = "teal")
plt.ylabel("Anomaly (million km2)")
plt.xlabel("Years")
plt.axhline(y=0.0, color = "orange")
plt.legend()

plt.twinx()

plt.plot(year3, sea_temp, color = "grey")
plt.ylabel("Sea Temperature (F)")
plt.xlabel("Years")

# assigning titles and values to the axes
plt.title("Sea Ice Anomaly Compared to Sea Temperature")
plt.xlim(year.min(), year.max())
plt.xticks(range(year.min(), year.max(), 5))
plt.legend()

# saving the graph
plt.savefig("Sea Ice Anomaly Compared to Sea Temperature.png")
plt.show()
```
